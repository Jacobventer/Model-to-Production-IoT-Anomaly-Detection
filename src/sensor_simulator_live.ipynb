```python
import time
import random
import requests
from datetime import datetime, timezone

api_url = "http://127.0.0.1:5000/predict"
machines = ["WT-01", "WT-02", "WT-03"]

while True:
    for machine in machines:
        payload = {
            "timestamp": datetime.now(timezone.utc).isoformat(),
            "machine_id": machine,
            "temperature": round(random.normalvariate(35, 4), 2),
            "humidity": round(random.normalvariate(45, 6), 2),
            "sound": round(random.normalvariate(60, 5), 2)
        }

        requests.post(api_url, json=payload)
        print(payload)

    time.sleep(1)
```
