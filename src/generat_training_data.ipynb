```python
#Create historical data for traing (offline)
import csv
import random
from datetime import datetime, timezone

output_file = "data/training_data.csv" #Save data in csv file
rows = 10000

#Noraml ranges
temp_range = (30, 45)
humidity_range = (20, 60)
sound_range = (50, 70)

with open(output_file, mode="w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["timestamp", "temperature", "humidity", "sound"])

    for _ in range(rows):
        writer.writerow([
            datetime.now(timezone.utc).isoformat(),
            round(random.uniform(*temp_range), 2),
            round(random.uniform(*humidity_range), 2),
            round(random.uniform(*sound_range), 2)
        ])

print(f"Training data written to {output_file}")
```
