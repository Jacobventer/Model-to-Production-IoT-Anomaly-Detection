```python
def rule_based_anomaly(data):
    if data["temperature"] > 45 or data["temperature"] < 5:
        return 1
    if data["humidity"] > 60 or data["humidity"] < 20:
        return 1
    if data["sound"] > 70 or data["sound"] < 40:
        return 1
    return 0
if __name__ == "__main__":
    test_sample = {
        "temperature": 55,
        "humidity": 30,
        "sound": 80
    }

    result = rule_based_anomaly(test_sample)
    print("Rule-based anomaly result:", result)

```
