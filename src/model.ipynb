```python
import joblib
import numpy as np

#Load trained model
model_path = "models/isolation_forest.pkl"
iso_forest = joblib.load(model_path)

#Feature order expected by trained model
feautures = ["temperature", "humidity", "sound"]

def predict_anomaly(sensor_data):
    #Convert JSON data to Numpy array
    X = np.array([[
        sensor_data["temperature"],
        sensor_data["humidity"],
        sensor_data["sound"]
    ]])

    prediction = iso_forest.predict(X)[0]
    score = iso_forest.decision_function(X)[0]

    anomaly_flag = 1 if prediction == -1 else 0 

    return anomaly_flag, float(score)

print("Done") #Confirms model loaded successfully
```
