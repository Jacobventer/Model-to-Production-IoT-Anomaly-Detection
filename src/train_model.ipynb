```python
#This script trains an Isolation Forest offline. It is not part of the real-time inderence API
import pandas as pd
from sklearn.ensemble import IsolationForest
import joblib
import os

#Paths
data_path = "data/training_data.csv"
model_path = "models/isolation_forest.pkl"

#Load training data from CSV
df = pd.read_csv(data_path)

#Select features for training
X = df[["temperature", "humidity", "sound"]]

#Train Isolation Forest
iso_forest = IsolationForest(
    n_estimators=100, #Number of Isolation tress
    contamination=0.03, #Expected anomaly rate
    random_state=42
)

iso_forest.fit(X)

#Model directory
os.makedirs("models", exist_ok=True)
#Save model as .pkl
joblib.dump(iso_forest, model_path)

print("Model trained and saved to", model_path)
```
